```python
import re
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file
with open("ChairV2.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Trích xuất thông tin bằng regex
top1_acc = re.findall(r'Top 1 accuracy:\s+([\d.]+)', text)
top5_acc = re.findall(r'Top 5 accuracy:\s+([\d.]+)', text)
top10_acc = re.findall(r'Top 10 accuracy:\s+([\d.]+)', text)
loss = re.findall(r'Loss:\s+([\d.]+)', text)

# Chuyển đổi dữ liệu sang dạng số
top1_acc = list(map(float, top1_acc))
top5_acc = list(map(float, top5_acc))
top10_acc = list(map(float, top10_acc))
loss = list(map(float, loss))

# Tạo danh sách epoch dựa trên số lượng loss
epochs = list(range(1, len(loss) + 1))

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(epochs, top1_acc, linestyle='-', label="Top 1 Accuracy", color='blue')
plt.plot(epochs, top5_acc, linestyle='-', label="Top 5 Accuracy", color='orange')
plt.plot(epochs, top10_acc, linestyle='-', label="Top 10 Accuracy", color='green')
plt.plot(epochs, loss, linestyle='-', label="Loss", color='red')

plt.xlabel("Epochs")
plt.ylabel("Value")
plt.title("Training progress on QMUL-Chair-V2")
plt.legend()
plt.grid(True)
plt.show()
```
